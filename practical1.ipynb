## 2) Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the 
## values entered are duplicates otherwise it prints "ALL UNIQUE".

```python
values = []
for i in range(6):
    values.append(int(input("Enter number: ")))

if len(values) != len(set(values)):
    print("DUPLICATES")
else:
    print("ALL UNIQUE")
```

### 4) Write a program which finds sum of digits of a number. 
### Example: n=130 then output is 4 (1+3+0).

```python
n = int(input("Enter a number: "))
sum_digits = 0

while n > 0:
    sum_digits += n % 10
    n //= 10

print("Sum of digits:", sum_digits)
```

### 1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'. 

```python
n = int(input("Enter n: "))

print("Prime numbers:")
for num in range(2, n + 1):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
```

### 4) Write a program to reverse a given number.

```python
n = int(input("Enter a number: "))
rev = 0

while n > 0:
    rev = rev * 10 + n % 10
    n //= 10

print("Reversed number:", rev)
```

### 1) Write a Sequential search function which searches an item in a sorted list. The function 
### should return the index of element to be searched in the list.

```python
def sequential_search(arr, key):
    for index in range(len(arr)):
        if arr[index] == key:
            return index
    return -1

arr = [10, 20, 30, 40, 50]
key = int(input("Enter element to search: "))

result = sequential_search(arr, key)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

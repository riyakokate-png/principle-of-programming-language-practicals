### 1) Write a program to count the number of vowels in a given string. 

```python
s = input("Enter a string: ")
count = 0

for ch in s:
    if ch.lower() in 'aeiou':
        count += 1

print("Number of vowels:", count)
```

### 2) Write a program to check if a string is a palindrome (reads the same forwards and 
### backwards). 

```python
s = input("Enter a string: ")
rev = s[::-1]

if s == rev:
    print("Palindrome")
else:
    print("Not a palindrome")
```

### 3) Write a program to reverse a string without using built-in reverse functions. 

```python
s = input("Enter a string: ")
rev = ""

for ch in s:
    rev = ch + rev

print("Reversed string:", rev)
```

### 5) Write a program to remove all whitespace from a string.

```python
s = input("Enter a string: ")
result = ""

for ch in s:
    if ch != " ":
        result += ch

print("String without whitespace:", result)
```

### 2) Write a program to check if two strings are anagrams (contain same characters in different 
## order).

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not anagrams")
```

### 3) Write a program to convert a string to title case without using the title() method. 
### Example: "python programming" should become "Python Programming"

```python
s = input("Enter a string: ")
words = s.split()
result = ""

for word in words:
    result += word[0].upper() + word[1:].lower() + " "

print("Title case string:", result.strip())
```

### 1) Write a program to find the longest word in a sentence.

```python
s = input("Enter a sentence: ")
words = s.split()

longest = words[0]
for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```

### 2) Write a program to check if a string contains only digits, only alphabets, or both. 

```python
s = input("Enter a string: ")

if s.isdigit():
    print("Only digits")
elif s.isalpha():
    print("Only alphabets")
else:
    print("Contains both digits and alphabets")
```

### 1) Write a program to find the union, intersection, and difference of two sets entered by the user.

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())

print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference (Set1 - Set2):", set1 - set2)
```

### 2) Write a program to check if a set is a subset of another set.

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())

if set1.issubset(set2):
    print("First set is a subset of second set")
else:
    print("First set is NOT a subset of second set")
```

### 3) Write a program to remove duplicates from a list using sets. 

```python
lst = input("Enter list elements: ").split()

unique_list = list(set(lst))

print("List after removing duplicates:", unique_list)
```

### 4) Write a program to find elements that are present in the first set but not in the second set. 

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())

result = set1 - set2

print("Elements only in first set:", result)
```

### 5) Write a program to check if two sets are disjoint (have no common elements).

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())

if set1.isdisjoint(set2):
    print("The sets are disjoint")
else:
    print("The sets are NOT disjoint")
```

### 1) Write a program to find symmetric difference between two sets.

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())

print("Symmetric Difference:", set1 ^ set2)
```

### 2) Write a program to add multiple elements to a set at once from user input.

```python
s = set()

elements = input("Enter elements to add: ").split()
s.update(elements)

print("Updated set:", s)
```

### 3) Write a program to find the length of a set and check if a specific element exists in it.

```python
s = set(input("Enter elements of the set: ").split())
element = input("Enter element to search: ")

print("Length of set:", len(s))

if element in s:
    print("Element exists in the set")
else:
    print("Element does not exist in the set")
```

### 1) Write a program that takes two sentences as input and finds: 
● Common words in both sentences 
● Words unique to the first sentence 
● Words unique to the second sentence 

```python
sentence1 = input("Enter first sentence: ").lower().split()
sentence2 = input("Enter second sentence: ").lower().split()

set1 = set(sentence1)
set2 = set(sentence2)

print("Common words:", set1 & set2)
print("Words unique to first sentence:", set1 - set2)
print("Words unique to second sentence:", set2 - set1)
```

### 2) Write a program to perform all set operations (union, intersection, difference, symmetric difference) on three sets.

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())
set3 = set(input("Enter elements of third set: ").split())

print("Union:", set1 | set2 | set3)
print("Intersection:", set1 & set2 & set3)
print("Difference (Set1 - Set2 - Set3):", set1 - set2 - set3)
print("Symmetric Difference:", set1 ^ set2 ^ set3)
```

## Question Set A
### 1. Create a dictionary with roll number as key and name as value for 5 students

```
students = {}

for i in range(5):
    roll = input("Enter roll number: ")
    name = input("Enter name: ")
    students[roll] = name

print("Student Dictionary:")
for k, v in students.items():
    print(k, ":", v)
```

### 2. Count the frequency of each word in a sentence

```
sentence = input("Enter a sentence: ").split()
freq = {}

for word in sentence:
    freq[word] = freq.get(word, 0) + 1

print(freq)
```

### 3. Merge two dictionaries

```
d1 = eval(input("Enter first dictionary: "))
d2 = eval(input("Enter second dictionary: "))

merged = d1 | d2
print("Merged Dictionary:", merged)
```

### 4. Find the key with the maximum value

```
d = eval(input("Enter dictionary: "))
max_key = max(d, key=d.get)
print("Key with maximum value:", max_key)
```

### 5. Check if a key exists in a dictionary

```
d = eval(input("Enter dictionary: "))
key = input("Enter key to search: ")

if key in d:
    print("Key exists")
else:
    print("Key does not exist")
```

## Question Set B
### 1. Remove a key and handle if it does not exist

```
d = eval(input("Enter dictionary: "))
key = input("Enter key to remove: ")

if key in d:
    del d[key]
    print("Updated dictionary:", d)
else:
    print("Key not found")
```

### 2. Sort dictionary by values in descending order

```
d = eval(input("Enter dictionary: "))
sorted_dict = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))
print(sorted_dict)
```

### 3. Create dictionary from two lists

```
keys = input("Enter keys: ").split()
values = input("Enter values: ").split()

d = dict(zip(keys, values))
print(d)
```

## Question Set C
### 1. Find all keys with the same value

```
d = eval(input("Enter dictionary: "))
value = input("Enter value to search: ")

keys = [k for k, v in d.items() if str(v) == value]
print("Keys with same value:", keys)
```

### 2. Create dictionary with numbers from 1 to n and their cubes

```
n = int(input("Enter n: "))
cube_dict = {i: i**3 for i in range(1, n+1)}
print(cube_dict)
```
